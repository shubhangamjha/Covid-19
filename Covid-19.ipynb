```python
df = df.drop(['death_reconsille', 'total', 'actualdeath24hrs'], axis=1)
df.head()
```

# **Approach**

```python
# Top seven active states
data_ative = df.sort_values(['active'], ascending=False)
data_ative = data_ative.iloc[:7, :]
data_ative
```

```python
explode = [0.1,0,0,0,0,0,0]
colors = ["r","b",'y','g','c','m','brown']
textprops = {"fontsize":8}
wedgeprops = {"linewidth":1, "width":1, "edgecolor":"black"}
fontdict = {"fontsize":10}
plt.title("top seven corona affected states \n Total Active cases in India: {} \n Total active cases in these states: {}%".format(sum(df['active']), round(sum(data_ative['active']),sum(df['active'], 4)*100), fontdict = fontdict))
plt.pie(data_ative['active'], labels=data_ative['state_name'], explode=explode, colors=colors, autopct="%0.2f", radius=1, wedgeprops=wedgeprops, textprops=textprops, pctdistance=0.6, labeldistance=1.1, rotatelabels=True)
plt.legend(loc=0)
plt.show()
```

# **Approach 2**

```python
def top_seven_states_index(x):
    positive = list(x)
    top_seven_states = []
    index_active = []
    state_active = []
    
    for i in range(7):
        top_seven_states.append(max(positive))
        positive.remove(max(positive))
        
    for j in range(len(top_seven_states)):
        for i in range(len(x)):
            if top_seven_states[j]==x[i]:
                index_active.append(i)
                
                
    for i in index_active:
        state_active.append(df['state_name'][i])
        
    for i in index_active:
        top_seven_states.append(x[i])
        
    return  top_seven_states, state_active, index_active
```

```python
top_seven_states, state_active, index_active = top_seven_states_index(df['active'])
```
